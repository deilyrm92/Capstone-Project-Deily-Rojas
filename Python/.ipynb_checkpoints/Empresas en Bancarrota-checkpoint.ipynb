### Revisión de Datos Faltantes

```python
total = df.isnull().sum().sort_values(ascending = False)
percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
pd.concat([total, percent], axis=1, keys=['Total', 'Percent']).transpose()
```

```python
df.dropna()
```

```python
missing_df_i = df.columns[df.isnull().any()].tolist()
msno.matrix(df[missing_df_i], figsize=(20,5))
```

```python
#Al ser alta la cantidad de Datos faltantes se debe proceder en remplazarlos
    # Construct an imputer with strategy as 'mean', to mean-impute along the columns
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer=imputer.fit(df)
dfNA=imputer.transform(df)
datasetf = pd.DataFrame(dfNA)
datasetf.columns=df.columns
datasetf
```

```python
datasetf.describe()
```

```python
#renombran las variables a float  
```

```python
index = 1
while(index<=63):
      colname = datasetf.columns[index]
      col = getattr(datasetf, colname)
      datasetf[colname] = col.astype(float)
      index+=1     
datasetf.info()
```

### Al ser un módelo de Clasificación, Se cambian variables a categoricas

```python
#Se cambia el tipo a categoria
datasetf=datasetf.astype({'class':'category'})
```

```python
#Rename Columna de Bancarrota por valores de mejor entendimiento
datasetf['class']=datasetf['class'].replace({0.0: "N"})
datasetf['class']=datasetf['class'].replace({1.0: "Y"})
```

```python
#dependent variable
depVar = datasetf['class']
#features
features = datasetf.iloc[:,0:44]
features
print('Summary of feature sample')
features.info()
```

# Split de información en Train y Test

```python
X_train, X_test,y_train, y_test = train_test_split(features,depVar)
```

```python
X_train.shape, X_test.shape  
```

```python
y_train.shape, y_test.shape
```

# Definición de Parametros de los Algortimos que se utilizaran para el modelado

```python
 ModelRF = RandomForestClassifier(n_jobs=4, 
                             random_state=20,
                             criterion='gini',
                             n_estimators=20,
                             verbose=False)

ModelADA=AdaBoostClassifier(random_state=20,
                             algorithm='SAMME.R',
                             learning_rate=0.8,
                             n_estimators=20)

ModelSVM=svm.SVC(C=1.0,
                verbose=False,
                random_state=20,
                )
```

# Random Forest

```python
ModelRF.fit(X_train, y_train)
```

```python
print(cross_val_score(ModelRF, X_train, y_train))
```

```python
ModelRF.score (X_train, y_train)
```

# AdaBoostClassifier

```python
ModelADA.fit(X_train, y_train)
```

```python
print(cross_val_score(ModelADA, X_train, y_train))
```

```python
ModelADA.score (X_train, y_train)
```

# SVM

```python
ModelSVM.fit(X_train, y_train)
```

```python
print(cross_val_score(ModelSVM, X_train, y_train))
```

```python
ModelSVM.score (X_train, y_train)
```

# Prediccion

```python
#Make Predictions
predictions = ModelRF.predict(X_test)
#print('RMSE: %.3f' % rmse)
m1=accuracy_score(y_test, predictions, normalize=True)
m2=cohen_kappa_score(y_test, predictions)
print('Accuracy: %.3f' % m1)
print('Kappa: %.3f' % m2)
```

```python
fig,ax=plt.subplots()
sns.distplot(y_test,
            hist=False, color='green',
            kde_kws={"shade":True},
            ax=ax, label='Real')
sns.distplot(predictions,
            hist=False, color='blue',
            kde_kws={"shade":True},
            ax=ax, label='Predict')
ax.set_title("Distribución por Default por Edad")
ax.legend(title='Default:')
```
